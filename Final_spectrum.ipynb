```python
import numpy as np
from astropy.io import ascii
from astropy.table import Table
import matplotlib.pyplot as plt
%matplotlib widget

# === GR7 File Paths ===
spec_path_gr7 = '/home/abhayprusty/Documents/projects/Exoplanets/Datafiles/2025-05-07PC2/cosmic_cleaned/bias_subtracted/flat_fielded/J16111534_modified/wlcspec_aie07056.txt'
irf_path_gr7 = '/home/abhayprusty/Documents/projects/Exoplanets/Datafiles/2025-05-07PC2/cosmic_cleaned/bias_subtracted/flat_fielded/stdstar/gr7_norm_con_IRF_wlcspec_aie07048.txt'  

# === Read GR7 Spectrum and IRF ===
spec7 = ascii.read(spec_path_gr7, format='fixed_width')
wl_spec7 = np.array(spec7['wavelength'])
flux_spec7 = np.array(spec7['flux'])
flux_err_spec7 = np.array(spec7['flux-err'])  

irf7 = ascii.read(irf_path_gr7, format='fixed_width')
wl_irf7 = np.array(irf7['Wavelength (A)'])
flux_irf7 = np.array(irf7['Flux'])
flux_err_irf7 = np.array(irf7['Flux-Err'])

# === Interpolate IRF to Spectrum ===
irf_interp7 = np.interp(wl_spec7, wl_irf7, flux_irf7)
irf_interp_err7 = np.interp(wl_spec7, wl_irf7, flux_err_irf7)

# === Correct Flux ===
flux_corrected7 = flux_spec7 / irf_interp7

rel_error_sq7 = (flux_err_spec7 / flux_spec7)**2 + (irf_interp_err7 / irf_interp7)**2
flux_corrected_err7 = flux_corrected7 * np.sqrt(rel_error_sq7)

# === Plot GR7 ===
plt.figure(figsize=(12, 4))
plt.plot(wl_spec7, flux_corrected7, label='GR7: IRF-Corrected Spectrum', color='blue')
plt.errorbar(wl_spec7, flux_corrected7, yerr=np.abs(flux_corrected_err7), color='k', alpha=0.3)
plt.xlabel('Wavelength ')
plt.ylabel('Flux')
plt.title('GR7: IRF-Corrected Spectrum with Error Propagation')
plt.legend()
plt.xlim(left=4200)
plt.ylim(0, 4000)
plt.tight_layout()
#plt.savefig("IRF_corrected_spectrum_gr7.pdf", format="pdf")
plt.show()

# === Save Output ===
output_gr7 = Table([wl_spec7, flux_corrected7, flux_corrected_err7],
                   names=('Wavelength', 'Corrected_Flux', 'Corrected_Flux_Error'))

save_path_gr7 = spec_path_gr7.replace('wlcspec_', 'IRF_corrected_')
#ascii.write(output_gr7, save_path_gr7, format='commented_header', overwrite=True)
print(f"GR7 IRF-corrected spectrum saved to:\n{save_path_gr7}")
```
